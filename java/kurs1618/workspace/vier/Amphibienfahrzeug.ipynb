```python
from graphviz import Digraph

g = Digraph('unix', filename='Amphibienfahrzeug.gv')
g.node_attr.update(color='lightblue2', style='filled')
g.body.append('size="6.6"')

g.node("Fahrzeug")
g.node("Auto")
g.node("Boot")
g.node("Amphibienfahrzeug")


g.edge("Fahrzeug", "Auto", dir="back")
g.edge("Fahrzeug", "Boot", dir="back")
g.edge("Auto", "Amphibienfahrzeug", dir="back")
g.edge("Boot", "Amphibienfahrzeug", dir="back")

g.view()
```
