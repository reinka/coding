### Here I don't want to restrict the number of layers but work with a more generalized version

```python
class NeuralNet(object):
    
    def __init__(self, sizes): 
        '''
        Initialize a NeuralNet object that takes as input the data matrix X, 
        the outcome vector y and an array containing the number of neurons
        in the respecitve layers of the network. [10, 5, 2] would be a three-layer
        network with the input layer consisting of 10 neurons, the hidden layer of 5
        and the output layer of 2 neurons.
        Based on the sizes, according weights will be created which are chosen
        randomly from a Gaussian distribution for symmetry breaking reasons.
        '''
        
        self.num_layers = len(sizes)
        self.sizes = sizes
        # We need Weight Matrices of the dim (dim 2nd layer, dim 1st layer), 
        # (dim 3rd layer, dim 2nd layer) ... , (n-th layer, dim n-1-th layer) 
        self.weights = [np.random.randn(y,x) for x,y in zip(sizes[:-1], sizes[1:])]
        # Bias for each layer except input layer
        self.biases = [np.random.randn(y,1) for y in sizes[1:]]
        self.costs = [] # list to store cost per iteration
        self.iterations = None
        
    def train(self, X, y, iterations, lambda_, alpha):
        y = self.one_hot_encoding(y)
        self.iterations = iterations
        for n in range(iterations):
            cost = self.update(X, y, lambda_, alpha)
            self.costs.append(cost)
            if n%25 == 0:
                print("Iteration {0} complete. Cost: {1}".format(n, cost))
        self.plot_costs()
            
    def update(self, X, y, lambda_,alpha):
        '''
        Update weights by applying gradient descent via backprop.
        '''
        m = len(y)
        #feedforward
        activations, zs = self.feedforward(X,y)
        nabla_b, nabla_w = self.backprop(y, zs, activations, lambda_)
        self.weights = [w-(alpha/m)*nabla_w
                        for w, nabla_w in zip(self.weights, nabla_w)]
        self.biases = [b-(alpha/m)*nabla_b.transpose()
                       for b, nabla_b in zip(self.biases, nabla_b)]
        return self.cost(y,activations[-1], lambda_)

            
    def feedforward(self, X, y):
        '''
        Feedforward. Returns `activations` vector and z-vector
        for each layer. Note that there is no z-vector for the
        first (input) layer.
        '''
        a = X
        activations = []
        zs = []
        for w,b in zip(self.weights,self.biases):
            activations.append(a)
        
            z = np.dot(a, w.transpose()) + b.transpose()
            zs.append(z)
            a = self.sigmoid(z)
            
        # end of loop
        activations.append(a) # append last activation matrix
    
        return activations, zs
        
            
    def backprop(self, y, z_list, activations, lambda_):
        '''
        Gradient descent via Backprop. 

        Starting with last layer and finishing at 2nd. We don't
        assume any errors regarding the 1st layer since that 
        one represents the data input.

        Indexing might seem confusing, since we loop from
        back to forth. 

        `z_list`: list containing the z-vectors layer by layer ordered from
        first hidden to output layer (2nd, 3rd, ..., output). Note that 
        there is no z-vector for the first layer = input. 

        `activation`: list containing the activation vectors, 
        ordered (input, a1, a2, ..., output). 


        Returns: 
        `nabla_b`: Computed Error Matrix.
        `nabla_w`: Computed Weight Matrix.
        '''
        nabla_b = [None for i in z_list] # empty list to store error matrices
        nabla_w = [None for i in z_list] # empty list to store weights' derivative

        output_error = activations[-1] - y # simply the difference between pred and output

        nabla_b[-1] = output_error
        nabla_w[-1] = np.dot(output_error.transpose(), activations[-2])

        # backprop up to (including) 2nd hidden layer
        # exclude z-vector from input layer since there we use
        # the activation vector to compute the error
        index = -1
        for z, w in zip(z_list[-2::-1], self.weights[-1::-1]):
            delta = np.dot(nabla_b[index],w) 
            delta *= self.sigmoid_gradient(z)

            index -= 1
            nabla_b[index] = delta
            nabla_w[index] = np.dot(delta.transpose(), activations[index - 1])
            index -= 1

        # Regularization
        for w,n_w in zip(self.weights, nabla_w):
            n_w += (lambda_ / len(y)) * w;

        return nabla_b, nabla_w

        
    def cost(self, y, hypothesis, lambda_):
        '''
        Calculate cost function for given data matrix `X`, 
        ouput vector `y` and regularization `lambda_` matrix.
        We need to sum over all training examples m, for each
        hypothesis 1...k, where k denotes the number of 
        different ouputs, in our case the digis 0...9 (k=10)
        '''
        m, K = y.shape
        # use Hadamard product instead of looping over K
        J = 1/m * np.sum(np.multiply(-1 * y,np.log(hypothesis))- np.multiply((1-y), np.log(1-hypothesis)))

        # Add regularization Term
        reg = sum([np.sum(np.sum(np.square(w), axis=1)) for w in self.weights])
        J += lambda_/(2*m) * reg
        return J 
        
        
    def one_hot_encoding(self, y):
        '''
        One hot encoding for given vector y. 
        This will be used to convert the digital representation [0...9] of the 
        ouput vector y into a 10-dimensional unit vector
        '''
        # map 10 to zero because of indexing
        y[y== 10] = 0
        
        n_values = np.max(y) + 1
        # Need to reshape mapped ouput since it'll be a 3D Matrix [len(y), 1, n_values]
        # We need a 2D Matrix of the dimensions (Size y) x (Number of digits) e.g. 5000x10 
        # if y consists of 5000 values with digits 1...9
        return np.eye(n_values)[y].reshape(len(y), n_values)       
        
    def sigmoid(self,z):
        '''
        Compute sigmoid function
        '''
        return 1.0 / (1.0 + np.exp(-z))
    
    def sigmoid_gradient(self,z):
        '''
        Derivative of sigmoid function
        '''
        return self.sigmoid(z)*(1-self.sigmoid(z))
    def plot_costs(self):
        import matplotlib.pyplot as plt
        #import seaborn as sns
        
        x = list(range(self.iterations))
        y = self.costs
        plt.plot(x,y)
        
        plt.xlabel("Iteration")
        plt.ylabel("J")
        plt.title("J values (=cost) against the number of the iteration")
        plt.show()
    
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def feedforward(X, y, lambda_):
    a = X
    activations = []
    zs = []
    for w,b in zip(weights,biases):
            activations.append(a)
            
            z = np.dot(a, w.transpose()) + b.transpose()
            zs.append(z)
            
            a = sigmoid(z)
    # end of loop
    activations.append(a) # append last activation matrix
    
    return activations, zs

a, zs = feedforward(X, y_temp, lambda_=0)
_ = [print(a.shape) for a in a]
print()
_ = [print(z.shape) for z in zs]

# output of the neural net (=last layer output)

a2, zs2 = feedforward(X, y_dummy, lambda_=0)
a2[1]
```

### Backprop

```python
Image("backprop.png")
```

```python
def backprop(y, zs, activations, lambda_):
    '''
    Gradient descent via Backprop. 
    
    Starting with last layer and finishing at 2nd. We don't
    assume any errors regarding the 1st layer since that 
    one represents the data input.
    
    Indexing might seem confusing, since we loop from
    back to forth. 
    
    `zs`: list containing the z-vectors layer by layer ordered from
    first hidden to output layer (2nd, 3rd, ..., output). Note that 
    there is no z-vector for the first layer = input. 
    
    `activation`: list containing the activation vectors, 
    ordered (input, a1, a2, ..., output). 
    
    
    Returns: 
    `nabla_b`: Computed Error Matrix.
    `nabla_w`: Computed Weight Matrix.
    '''
    m = len(y)
    nabla_b = [None for i in zs] # empty list to store error matrices
    nabla_w = [None for i in zs] # empty list to store weights' derivatives
    output_error = activations[-1] - y # simply the difference between pred and output
    nabla_b[-1] = output_error
    nabla_w[-1] = np.dot(output_error.transpose(), activations[-2])
    # backprop up to (including) 2nd hidden layer
    # exclude z-vector from input layer since there we use
    # the activation vector to compute the error
    index = -1
    for z, w in zip(zs[-2::-1], weights[-1::-1]):
        delta = np.dot(nabla_b[index],w) 
        delta *= sigmoid_gradient(z)
        
        index -= 1
        nabla_b[index] = delta
        nabla_w[index] = np.dot(delta.transpose(), activations[index - 1])
        index -= 1
        
    # Regularization. Only for weights! 
    for w,n_w in zip(weights, nabla_w):
        n_w += (lambda_ / m) * w;
        
        
    return nabla_b, nabla_w
    
comp_backprop = backprop(y_dummy, zs2, a2, lambda_=1)
_ = [print(i.shape) for i in comp_backprop[1]]
comp_backprop[1][:1]
```

```python
def predict(weights, X):
    '''
    Predicts Label of input X for given weights
    '''
    x = 0
    for w in weights:
        x = np.insert(X, 0, 1, axis=1)
        z = np.dot(x, w.transpose)
        h = sigmoid(z)
        x = h
    
    # Index of max Value == digit
    return np.argmax(x)
```
