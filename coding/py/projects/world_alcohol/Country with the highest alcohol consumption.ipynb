```python
#Now choose only those rows, which include 'All types' since we want the total consumption of alcohol
sorted_world_alcohol = world_alcohol[world_alcohol['Beverage Type'] == 'All types']
sorted_world_alcohol
```

```python
#now we will sort for the years only, since they conain the values we need to calculate with
years_only = sorted_world_alcohol.iloc[:,2:]
years_only = years_only.fillna(0)
years_only.head()
```

```python
years_only.convert_objects(convert_numeric=True).dtypes
```

```python
#conerting all data to numeric
years_only = years_only.convert_objects(convert_numeric=True)
```

```python
#we create a new column, which contains the mean over all the years
average_consumption = years_only.mean(axis=1)
average_consumption.head()
```

```python
max_value = years_only.max()
years_only[years_only.]
```

```python
#add a new column with our aveage conssumption to the already sorted list (since row number has to match)
sorted_world_alcohol['Average consumption'] = average_consumption
sorted_world_alcohol
```

```python
#we got our country
#Method 1: sort
sorted_world_alcohol.sort('Average consumption', ascending=False).head()
```

```python
#Method 2: 
#find index and thus country which belongs to the max of average consumption
sorted_world_alcohol.Average[sorted_world_alcohol.Average == sorted_world_alcohol.Average.max()].index
```
