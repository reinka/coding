```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
p = 1/6
n = 100000
```

```python
outcome = np.random.randint(6, size=n)
```

```python
one = filter(lambda x: x == 0, outcome)
```

```python
successfull_rolls = len(list(one))
```

```python
proportion = []
n = 1
hits = 0
for i in outcome:
    if i == 0:
        hits += 1
    proportion.append(hits/n)
    n+=1        
```

```python
plt.figure(figsize=(2560/100, 1600/100), dpi=100)

plt.plot([i for i in range(1,n)], proportion)
plt.plot([i for i in range(1,n)],[1/6 for i in range(1,n)], 'r--')
plt.plot(1/6)


plt.ylim(-0.01,)
plt.xscale('log')
plt.title('Expected value of die rolls that are 1')
plt.xlabel('n (number of rolls)')
plt.ylabel('$\mathrm{\hat{p}_{n}}$')
plt.savefig('Expected value of die rolls that are 1', dpi = 100)
```
