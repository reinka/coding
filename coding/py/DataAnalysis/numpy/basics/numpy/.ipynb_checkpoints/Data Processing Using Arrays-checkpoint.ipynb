```python
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#numpy arrays enable to express data processing tasts as concise array
#expression that might otherwise require loops -> vectorization
```

```python
points = np.arange(-5,5,0.01) #1000 equally spaced points
```

```python
#np.meshgrid takes two 1D arrays and returns two 2D matrices 
#correspondening to all pairs of (x,y) in the two arrays:

xs,ys = np.meshgrid(points,points)
```

```python
ys
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
z
```

```python
plt.imshow(z); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
plt.savefig("meshgrid_plot.png")
```
