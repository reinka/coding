```python
import numpy as np
from numpy.random import randint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#in python:
import random

position = 0
walk = [position]
steps = 1000
for i in xrange(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk)
```

```python
#in numpy

nsteps= 1000
draws = randint(0,2,size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
(np.abs(walk)>=10).argmax()
```

```python
plt.title("Random walk")
plt.plot(walk)
```

```python
nwalks = 5000
nsteps = 1000
draws = randint(0,2, size=(nwalks,nsteps))
steps = np.where(draws >0,1,-1)
walks = steps.cumsum(1)
walks
plt.plot(walks)
```

```python
walks.min()
```

```python
walks.max()
```

```python
hits30 = (np.abs(walks) >= 30).any(1)
hits30
```

```python
hits30.sum()
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)
```

```python
crossing_times
```

```python
crossing_times.mean()
```
