### Part A

    

```python
import re

# From the google babynames.py file:

"""
Here's what the html looks like in the baby.html files:
...
<h3 align="center">Popularity in 1990</h3>
....
<tr align="right"><td>1</td><td>Michael</td><td>Jessica</td>
<tr align="right"><td>2</td><td>Christopher</td><td>Ashley</td>
<tr align="right"><td>3</td><td>Matthew</td><td>Brittany</td>
...

Suggested milestones for incremental development:
 -Extract the year and print it
 -Extract the names and rank numbers and just print them
 -Get the names data into a dict and print it
 -Build the [year, 'name rank', ... ] list and print it
 -Fix main() to use the extract_names list
"""
```

    Filter out names only

```python
f = open('/Users/slackoverflow/Downloads/google-python-exercises/babynames/baby2006.html')
names = re.findall(r'([a-z]+)</td>',f.read(),re.IGNORECASE)
names
```


    Create Hastable with rank = key and according names = value
    For this filter out the names with the according year first

```python
f = open('/Users/slackoverflow/Downloads/google-python-exercises/babynames/baby2006.html')

year_name_tuple = re.findall(r'(\d+)</td><td>(\w+)</td><td>(\w+)</td>', f.read())

# Hash table aka dict now
d = dict()
for year, name1, name2 in year_name_tuple:
    d[year] = [name1, name2]

d
```

    Sort names and get the according rank via the hash table.

```python
names_sorted = sorted(names)
hash_table = d.copy() # for improved reading

result = []
for name in names_sorted:
    for key, val in hash_table.items():
        if name in val:
            result += [name + ' ' + key]
            break
result
```

    Filter out the year and print the result.

```python
f = open('/Users/slackoverflow/Downloads/google-python-exercises/babynames/baby2006.html')
year = re.search(r'(\d\d\d\d)</h3>', f.read()).group(1)


print(year)
for item in result:
    print(item)
```
