```python
from lxml import html
#from urllib.request import Request, urlopen
import requests, bs4, re
```

### Extract URL

```python
url = "https://foundational-research.org/artificial-intelligence-and-its-implications-for-future-suffering"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
page = requests.get(url, headers=headers)
page.raise_for_status()
soup = bs4.BeautifulSoup(page.text, 'lxml')
```

### Get links and wirte an update function, which updates the container of the links, once some links have been filtered out

```python
box = []
for a in soup.find_all('a', href=True):
    print("Found the URL:", a['href'])
    box.append(a['href'])

def remove_from_box(box, remove):
    '''
    Function to update box once new urls have been 
    filtered out
    '''
    return [i for i in box if i not in remove]
```

### Remove links like footnotes, ToC, etc.

```python
box = [i for i in box if not i.startswith('#') and not i.startswith('/')]
```

### Count number of links and print each one

```python
print("Number of links: {0}".format(len(box)))
_ = [print(url) for url in box]
```

### PDFs will be Papers for sure, so filter them out

```python
print("PDFs:")
_ = [print(i) for i in box if i.endswith("pdf")]
box = remove_from_box(box, [i for i in box if i.endswith("pdf")])
print("URLs left:",len(box))
```

### Filter out links from wikipedia, youtube, facebook, utilitrian-essays, foundational-research etc. these ones won't contain any scientific papers

```python
box = remove_from_box(box, [i for i in box if i.startswith("https://en.wikipedia.")])
box = remove_from_box(box, [i for i in box if i.startswith("http://en.wikipedia.")])
box = remove_from_box(box, [i for i in box if i.startswith('https://foundational-research')])
box = remove_from_box(box, [i for i in box if i.startswith('https://www.facebook.')])
box = remove_from_box(box, [i for i in box if i.startswith("https://www.youtube")])
box = remove_from_box(box, [i for i in box if i.startswith("http://reducing-suffering")])
box = remove_from_box(box, [i for i in box if i.startswith("http://www.utilitarian-essays")])
box = remove_from_box(box, [i for i in box if i.startswith("https://www.utilitarian-essays")])
box = remove_from_box(box, [i for i in box if i.startswith("http://wiki.lesswrong")])
box = remove_from_box(box, [i for i in box if i.startswith("http://lesswrong")])




print("URLs left:",len(box))
```
